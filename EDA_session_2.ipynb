**create data frame using dictionary**

```python
name=['avinash','akash','adhya']
age=[20,25,30]
pd.DataFrame(zip(name,age),columns=['name','age'])
```

```python
dict={'Names':['avinash','akash','adhya'],
     'age':[20,25,30]}
pd.DataFrame(dict)

# keys becomes columns
# values becomes rows
# No need to provide columns names seperately
```

```python
dict2={'Name':'sagar','age':24,'city':'Hyd'}
pd.DataFrame(dict2,index=[1])
```

```python
dict2={'Name':'sagar','age':24,'city':'Hyd'}
pd.DataFrame(dict2,index=[1,2,3])
```
